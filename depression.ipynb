```python
depression_num = depression[depression_num_attribs]
```

```python
#Create a data pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

cat_attribs = demographic_qs

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])

depression_num_tr = num_pipeline.fit_transform(depression_num)

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, depression_num_attribs),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_attribs)
])

depression_prepared = full_pipeline.fit_transform(depression)
```

```python
#The predictions are so precise, we may have an overfitting problem. Solutions for overfitting are: (1) Simplify the model; (2) Constrain it (regularize it); (3) Get a lot more training data.
#We need to use part of the training set for training, and part of it for validation using cross_val_score.
```

```python
from sklearn.tree import DecisionTreeRegressor
dep_tree = DecisionTreeRegressor()
dep_tree.fit(depression_prepared, depression_rating)

dep_predictions = dep_tree.predict(depression_prepared)
tree_rmse = mean_squared_error(depression_rating, dep_predictions, squared = False)
tree_rmse
```

```python
#Again, we have a small rmse. We should simplify the model and validate the model on the training data.
depression_num_attribs_simplified = ['TIPI1', 'TIPI2',
       'TIPI3', 'TIPI4', 'TIPI5', 'TIPI6', 'TIPI7', 'TIPI8', 'TIPI9', 'TIPI10',
       'VCL1', 'VCL2', 'VCL3', 'VCL4', 'VCL5', 'VCL6', 'VCL7', 'VCL8', 'VCL9',
       'VCL10', 'VCL11', 'VCL12', 'VCL13', 'VCL14', 'VCL15', 'VCL16',
       'education', 'age', 'familysize']

cat_attribs = demographic_qs

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, depression_num_attribs_simplified),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_attribs)
])

depression_prepared = full_pipeline.fit_transform(depression)


#Linear regression
linreg = LinearRegression()
linreg.fit(depression_prepared, depression_rating)

#Now, get the rmse
dep_predictions =linreg.predict(depression_prepared)
lin_rmse = mean_squared_error(depression_rating, dep_predictions, squared = False)
lin_rmse
```

```python
depression_prepared.shape
```

```python
#The model is not overfitting now. As the depression score is from 0 to 1, this is not too bad. Let's try the decision tree.
dep_tree = DecisionTreeRegressor()
dep_tree.fit(depression_prepared, depression_rating)

dep_predictions = dep_tree.predict(depression_prepared)
tree_rmse = mean_squared_error(depression_rating, dep_predictions, squared = False)
tree_rmse
```

```python
#The decision tree RMSE is still 0, showing overfitting.
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dep_tree, depression_prepared, depression_rating, scoring = 'neg_mean_squared_error', cv = 10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print('scores:', scores)
    print('mean:', scores.mean())
    print('standard deviation:', scores.std())

display_scores(tree_rmse_scores)
```

```python
#Using folds, our scores worsened. Let's try random forests
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(max_features='sqrt')
forest_reg.fit(depression_prepared, depression_rating)
dep_tree_predictions = forest_reg.predict(depression_prepared)
forest_rmse = mean_squared_error(depression_rating, dep_tree_predictions, squared = False)
forest_rmse
```

```python
results = pd.DataFrame({'Predicted': dep_tree_predictions[:10].tolist(), 'Actual': depression_rating[:10].tolist()})
results['Error'] = abs(results['Predicted'] - results['Actual'])/results['Actual']
results.style.format({'Error': '{:,.2%}'.format,})
```

```python
x_test = test_set.drop('depression_rating', axis = 1)
x_test = x_test.drop('depression_score', axis = 1)
x_test = x_test.drop('major', axis = 1)
y_test = test_set['depression_rating'].copy()
```

```python
x_test.head()
```

```python
x_test_prepared = full_pipeline.transform(x_test)
```

```python
final_predictions = forest_reg.predict(x_test_prepared)
final_rmse = mean_squared_error(y_test, final_predictions, squared = False)
```

```python
final_mae = mean_absolute_error(y_test, final_predictions)
errors = abs(final_predictions - y_test)
mape = 100 * (errors / y_test)
accuracy = 100 - np.mean(mape)
print('Final root mean squared error:', round(final_rmse, 3))
print('Final mean absolute error:', round(final_mae, 3))
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
#Compute a 95% confidence interval
from scipy import stats
confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1, loc = squared_errors.mean(), scale = stats.sem(squared_errors)))
```
